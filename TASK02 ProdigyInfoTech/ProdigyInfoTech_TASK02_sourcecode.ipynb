# Visualize and Explore the data.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of survival rates
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Survived')
plt.title('Distribution of Survival Rates')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()

# Analyzing trends based on PassengerId
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='PassengerId', y='Survived', hue='Survived', palette='viridis', s=100)
plt.title('Survival by PassengerId')
plt.xlabel('PassengerId')
plt.ylabel('Survived')
plt.show()
```
