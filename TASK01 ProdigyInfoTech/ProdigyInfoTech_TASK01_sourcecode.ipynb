TASK#1: Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
#1) Analyze the dataset abd check total columns and rows.

import pandas as pd

dataset = pd.read_csv("Metadata_Country_API_SP.POP.TOTL_DS2_en_csv_v2_386866.csv")

print("The dataset is as follows: ")
dataset
```

```python
# Data Cleaning

# Checking for missing values
dataset.isnull().sum()
```

```python
# As per above output, we analyze that there are many missing values in many columns


# Fill missing values in Region and IncomeGroup with "Unknown" for visualization purposes.
dataset["Region"].fillna("Unknown", inplace=True)
dataset["IncomeGroup"].fillna("Unknown", inplace=True)
dataset["SpecialNotes"].fillna("Unknown", inplace=True)
dataset["Unnamed: 5"].fillna("Unknown", inplace=True)

dataset
```

```python
# Count the occurrences of each region

region_counts = dataset["Region"].value_counts()

region_counts
```

```python
# Count the occurrences of each IncomeGroup

income_counts = dataset["IncomeGroup"].value_counts()

income_counts
```

```python
#Create bar charts for the "Region" and "IncomeGroup" variables.
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting the distribution of regions
plt.figure(figsize=(12, 6))
region_counts.plot(kind="bar", color="lightgreen")
plt.title("Visualization of Distribution of Regions")
plt.xlabel("Region")
plt.ylabel("Number of Countries")
plt.xticks(rotation=90)
plt.show()
```

```python
# Plotting the distribution of IncomrGroup
plt.figure(figsize=(12, 6))
income_counts.plot(kind="bar", color="Orange")
plt.title("Visualization of Distribution of IncomeGroup")
plt.xlabel("IncomrGroup")
plt.ylabel("Number of Countries")
plt.xticks(rotation=90)
plt.show()

# Uncknown showing missing values
```
